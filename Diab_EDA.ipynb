```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
filepath='diabetes_dataset.csv'
def load_data(file_path):
    df = pd.read_csv(file_path)
    return df

if __name__ == "__main__":
    df = load_data(filepath)
df.head()
```

```python
# To know the Statistics of the Data
df.describe()
```

```python
# Get the correlation metrix of the Data
df.corr()
```

```python
# Correlation matrix
corr_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
#plt.savefig('Correlation matrix DB')
plt.show()
```

```python


# Histograms for numerical variables
df.hist(figsize=(15, 10), bins=20)
#plt.savefig(' Histograms for DB')
plt.show()

```

```python
# Boxplots for numerical variables
plt.figure(figsize=(15, 10))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar plot for categorical variables
plt.figure(figsize=(10, 6))
sns.countplot(x='Outcome', data=df)  # Replace 'Outcome' with other categorical columns
#plt.savefig('Bar plot')
plt.show()
```

```python
# Scatter plot between Age and Glucose
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Glucose', hue='Outcome', data=df)
#plt.savefig('Scatter plot DB')
plt.show()
```

```python
# Boxplot of BMI by Outcome
plt.figure(figsize=(8, 6))
sns.boxplot(x='Outcome', y='BMI', data=df)
#plt.savefig('Box plot BMI')
plt.show()
```

```python
# Cross-tabulation of FamilyHistory and Outcome
print(pd.crosstab(df['FamilyHistory'], df['Outcome']))

```

```python
# Grouped bar plot
sns.countplot(x='FamilyHistory', hue='Outcome', data=df)
#plt.savefig('Cross-tabulation')
plt.show()
```

```python
# Pair plot for selected variables
sns.pairplot(df[['Age', 'Glucose', 'BMI', 'Outcome']], hue='Outcome')
#plt.savefig('pari plot')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Age'], df['Glucose'], df['BMI'], c=df['Outcome'], cmap='viridis')
ax.set_xlabel('Age')
ax.set_ylabel('Glucose')
ax.set_zlabel('BMI')
#plt.savefig('3D')
plt.show()
```

```python
# Boxplot to detect outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Age', 'Glucose', 'BMI']])
#plt.savefig('Bar plot outlier detection')
plt.show()
```

```python
# Z-score method to detect outliers
from scipy.stats import zscore
df['z_score'] = zscore(df['Glucose'])
outliers = df[df['z_score'].abs() > 3]
print("Outliers in Glucose:")
outliers
```

```python
# Calculate mean and std of Glucose
glucose_mean = df['Glucose'].mean()
glucose_std = df['Glucose'].std()

# Save to a file
np.savez('glucose_stats.npz', mean=glucose_mean, std=glucose_std)
```

```python
# Create BMI categories
df['BMI_Category'] = pd.cut(df['BMI'], bins=[0, 18.5, 24.9, 29.9, 100], labels=['Underweight', 'Normal', 'Overweight', 'Obese'])

# Create Age groups
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 30, 50, 100], labels=['18-30', '31-50', '51+'])

# Create Glucose categories
df['Glucose_Category'] = pd.cut(df['Glucose'], bins=[0, 99, 125, 300], labels=['Normal', 'Prediabetes', 'Diabetes'])
```

```python
# One-hot encoding for categorical variables
df = pd.get_dummies(df, columns=['BMI_Category', 'Age_Group', 'Glucose_Category'], drop_first=True)
```

```python
# Create BMI categories
df['BMI_Category'] = pd.cut(df['BMI'], bins=[0, 18.5, 24.9, 29.9, 100], labels=['Underweight', 'Normal', 'Overweight', 'Obese'])
# Set the style
sns.set(style="whitegrid")

# Plot BMI Category distribution
plt.figure(figsize=(8, 6))
sns.countplot(x=df['BMI_Category'], data=df, palette='viridis')
plt.title('BMI Category Distribution')
plt.xlabel('BMI Category')
plt.ylabel('Count')
plt.show()
```

```python
# Create Age groups
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 30, 50, 100], labels=['18-30', '31-50', '51+'])
# Plot Age Group distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Age_Group', data=df, palette='magma')
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
# Create Glucose categories
df['Glucose_Category'] = pd.cut(df['Glucose'], bins=[0, 99, 125, 300], labels=['Normal', 'Prediabetes', 'Diabetes'])
# Plot Glucose Category distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Glucose_Category', data=df, palette='plasma')
plt.title('Glucose Category Distribution')
plt.xlabel('Glucose Category')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter plot of BMI vs Glucose
plt.figure(figsize=(8, 6))
sns.scatterplot(x='BMI', y='Glucose', hue='Glucose_Category', data=df, palette='coolwarm')
plt.title('BMI vs Glucose')
plt.xlabel('BMI')
plt.ylabel('Glucose')
plt.legend(title='Glucose Category')
plt.show()
```

```python
# Box plot of BMI by Age Group
plt.figure(figsize=(8, 6))
sns.boxplot(x='Age_Group', y='BMI', data=df, palette='Set2')
plt.title('BMI Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('BMI')
plt.show()
```
